# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)

## Objective
The goal of this task is to load, inspect, and visualize the Iris dataset to understand data trends and feature distributions.

## Dataset
Iris Dataset (loaded using seaborn)


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df
```

```python
print("Shape of Dataset:",df.shape)
```

```python
print("Columns:",df.columns)
```

```python
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=df)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
df.hist(figsize=(10,8))
plt.suptitle("Histogram of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Box Plot of Seal Length")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Box Plot of Peal Length")
plt.show()
```

```python
sns.pairplot(df ,hue="species")
plt.show()
```

# Conclusion / Final Insights

- The Iris dataset contains 150 rows and 5 columns.
- The dataset has no missing values.
- Scatter plots show relationships between features, especially petal length and petal width.
- Histograms show how values are distributed across different features.
- Box plots help identify possible outliers and compare feature ranges across species.
